```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
swarm1 = np.array(pd.read_csv('oswarms/oswarm0.csv'))
swarm2 = np.array(pd.read_csv('oswarms/oswarm2.csv'))
```

```python
for part in swarm1:
    part_xy = part.reshape((50,2))
    plt.scatter(part_xy[:,0], part_xy[:,1], alpha=0.1, c='blue')
for part in swarm2:
    part_xy = part.reshape((50,2))
    plt.scatter(part_xy[:,0], part_xy[:,1], alpha=0.1, c='red')
```
